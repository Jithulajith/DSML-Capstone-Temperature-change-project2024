### 12. Model Evaluation:

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score
import numpy as np

from sklearn.svm import SVR
from sklearn.neural_network import MLPRegressor
from sklearn.ensemble import RandomForestRegressor, GradientBoostingRegressor, AdaBoostRegressor
from sklearn.linear_model import LinearRegression

models = {
    'SVR': SVR(C=10, gamma='scale', kernel='rbf'),
    'MLP Regressor': MLPRegressor(hidden_layer_sizes=(50,), activation='relu', max_iter=500, random_state=42),
    'Random Forest Regressor': RandomForestRegressor(n_estimators=100, max_depth=20, random_state=42),
    'Gradient Boosting Regressor': GradientBoostingRegressor(n_estimators=100, learning_rate=0.1, random_state=42),
    'AdaBoost Regressor': AdaBoostRegressor(n_estimators=50, learning_rate=0.01, random_state=42),
    'Linear Regression': LinearRegression()
}

results = {}
for name, model in models.items():
  
    model.fit(X_train, y_train)
    
 
    y_pred = model.predict(X_test)
    
    mae = mean_absolute_error(y_test, y_pred)
    mse = mean_squared_error(y_test, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_test, y_pred)
    
   
    results[name] = {
        'MAE': mae,
        'MSE': mse,
        'RMSE': rmse,
        'R2 Score': r2
    }


for name, metrics in results.items():
    print(f"\n{name}:")
    print(f"MAE: {metrics['MAE']:.4f}")
    print(f"MSE: {metrics['MSE']:.4f}")
    print(f"RMSE: {metrics['RMSE']:.4f}")
    print(f"R2 Score: {metrics['R2 Score']:.4f}")
```

**Model Evaluation: Trained and evaluated the models using metrics such as MAE, MSE, RMSE, and R² score to determine their performance.**

### Dataset Limitations

1. **Feature Diversity**: The dataset only includes temperature change as the target, without additional contextual variables like humidity, CO₂ levels, or industrial activity, which could help the models make more accurate predictions.
  
2. **Temporal Granularity**: The dataset contains annual data, but temperature can fluctuate significantly on a monthly or seasonal basis. More granular data could provide deeper insights.

3. **Geographical Representation**: The dataset may have some countries overrepresented, which could bias model predictions, particularly when generalizing to less-represented regions.

4. **Imputation of Missing Data**: Missing data was handled through imputation, which, while necessary, could introduce errors if the missing data were not random or properly handled.

5. **Lack of External Factors**: Including factors like economic growth, urbanization, or deforestation rates could significantly improve the predictive capability of the models by accounting for other drivers of temperature change.

### Conclusion

This analysis demonstrates that the **Gradient Boosting Regressor** and **Random Forest Regressor** offer the most accurate predictions for temperature changes, outperforming other models in terms of error metrics and explanatory power (R²). However, the limitations in the dataset suggest that there is room for improvement.

For future work, it is recommended to:
